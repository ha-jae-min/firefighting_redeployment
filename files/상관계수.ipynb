```python
import pandas as pd

# df = pd.read_excel('/content/b21_kor_V2.5.xlsx')
# df = pd.read_excel('/content/b21_kor_V2.75.xlsx')
df = pd.read_excel('/content/b2.xlsx')
df
```

```python
column_names = df.columns.tolist()
print(column_names)
```

```python
# 데이터프레임에서 모든 컬럼을 선택
all_columns = df.columns

# 상관계수를 저장할 데이터프레임 초기화
correlation_matrix = pd.DataFrame(index=all_columns, columns=all_columns)

# 모든 컬럼 쌍에 대한 상관계수 계산
for col1 in all_columns:
    for col2 in all_columns:
        correlation = df[col1].corr(df[col2])
        correlation_matrix.loc[col1, col2] = correlation

# 결과 출력
correlation_matrix
```

```python
# 양수의 상관계수를 가지는 컬럼 쌍을 추출
positive_correlation_pairs = []
for col1 in all_columns:
    for col2 in all_columns:
        if col1 != col2:
            correlation = float(correlation_matrix.loc[col1, col2])
            if correlation >= 0.3:
                positive_correlation_pairs.append((col1, col2, correlation))

# 결과 출력
for pair in positive_correlation_pairs:
    col1, col2, correlation = pair
    print(f"Positive correlation between {col1} and {col2}: {correlation}")
```

```python
selected_pairs = []

for pair in positive_correlation_pairs:
    col1, col2, correlation = pair
    if '출동소요시간_scld' in (col1, col2) or '재산피해금액_scld' in (col1, col2) or '화재진압시간_scld' in (col1, col2):
        selected_pairs.append((col1, col2, correlation))

# 결과 출력
for pair in selected_pairs:
    col1, col2, correlation = pair
    print(f"Positive correlation between {col1} and {col2}: {correlation}")
```
